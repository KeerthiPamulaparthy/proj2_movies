# Code to join data from Boxoffice mojo and The numbers

```python
import pandas as pd
```

```python
movies_boxoffice = pd.read_csv('boxofficemojo_animation.csv')
movies_boxoffice = movies_boxoffice.drop('Unnamed: 0', axis = 1)
```

```python
movies_boxoffice.head()
```

```python
movies_numbers = pd.read_csv('numbersbudgets.csv')
movies_numbers.head()

```

```python
movies_numbers = movies_numbers.drop('release_date', axis = 1)
```

```python
movies_numbers.head()
```

```python
movies_result = pd.concat([movies_boxoffice, movies_numbers], axis=1, join_axes=[movies_boxoffice.index])
movies_result.head()
```

```python
import difflib
movies_boxoffice['movie_name'] = movies_boxoffice['movie_name'].apply(lambda x: difflib.get_close_matches(x, movies_numbers['movie_name'],n=1,cutoff=.9)[0] if difflib.get_close_matches(x, movies_numbers['movie_name'],n=1,cutoff=.9) else x)

 
```

```python
movies_result_v5 = pd.merge(movies_boxoffice, movies_numbers, how ='left', on='movie_name')
movies_result_v5.head()
```

```python
movies_result_v5['production_budget_x'].fillna(movies_result_v5['production_budget_y'], inplace=True)
movies_result_v5.head()
```

```python
movies_result_v5 = movies_result_v5.drop(['production_budget_y'],axis = 1)
```

```python
movies_result_v5.head()
```

```python
movies_result_v5 = movies_result_v5.drop(['Unnamed: 0'],axis = 1)
```

```python
all_movies = movies_result_v5
```

```python
all_movies.to_csv('animation_movies.csv')
```
